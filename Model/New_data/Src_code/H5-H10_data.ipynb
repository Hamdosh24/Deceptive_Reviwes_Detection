```python
def perceptual_cognitive(text, perceptual_phrases):
  for phrase in perceptual_phrases:
    if phrase in text:
      return True
  return False
```

# H5

```python
def get_h5_reviews(*data):
  h5_reviews = []
  for reviews in data:
    for _, row in reviews.iterrows():
      text = emoji.replace_emoji(str(row["text"]), replace='')
      res = perceptual_cognitive(text, perceptual_phrases)
      if not res:
        h5_reviews.append(row["text"])
  return h5_reviews
```

```python
h5_reviews = get_h5_reviews(shein_data, afrd_data, company_data, balance_reviews)
```

# H10

```python
MINIMUM_LENGTH = 10

def get_h10_reviews(*data):
  h10_reviews = []
  for reviews in data:
    for _, row in reviews.iterrows():
      text = emoji.replace_emoji(str(row["text"]), replace='')
      unique_words = set(text.split())
      if len(unique_words) > MINIMUM_LENGTH and 'نقاط' not in unique_words:
        res = perceptual_cognitive(text, perceptual_phrases)
        if res:
          h10_reviews.append(row["text"])

  return h10_reviews
```

```python
h10_reviews = get_h10_reviews(shein_data, afrd_data, company_data, balance_reviews)
```

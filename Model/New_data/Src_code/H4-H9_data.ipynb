```python
shein_data = pd.read_excel("/content/drive/MyDrive/New_data/unlabeld_data/Arabic_Reviews_of_SHEIN.xlsx")
afrd_data = pd.read_excel("/content/drive/MyDrive/New_data/unlabeld_data/AFRD_multi-domain.xlsx")
company_data = pd.read_csv("/content/drive/MyDrive/New_data/unlabeld_data/Company_Reviews.cvs")
balance_reviews = pd.read_excel("/content/drive/MyDrive/New_data/unlabeld_data/Balance_reviews.xlsx")
```

```python
cognitive_phrases = [
    "اظن",
    "اتوقع",
    "بظن",
    "طنينا"
    "احس",
    "حسينا"
    "فكرت",
    "فكرنا"
    "انصح",
    "ننصح",
    "اعرف",
    "بعرف",
    "نعرف",
    "سمعت",
    "سمعنا",
    "جيبه",
    "جيبوه",
    "اشتريه",
    "اشتروه",
    "احجزه"
    "احجزوه"
    "دائما",
    "ابدا",

]
```

```python
def contains_cognitive(text, cognitive_phrases):
  for phrase in cognitive_phrases:
    if phrase in text:
      return True
  return False
```

# H4

```python
def get_h4_reviews(*data):
  h4_reviews = []
  for reviews in data:
    for _, row in reviews.iterrows():
      text = emoji.replace_emoji(str(row["text"]), replace='')
      res = contains_cognitive(text, cognitive_phrases)
      if res:
        h4_reviews.append(row["text"])
  return h4_reviews
```

```python
h4_reviews = get_h4_reviews(shein_data, afrd_data, company_data, balance_reviews)
```

```python
len(h4_reviews)
```

```python
h4_df = pd.DataFrame(h4_reviews, columns=["Text"])
h4_df.to_csv("/content/drive/MyDrive/New_data/Data/h4_reviews.csv", index=False)
```

# H9

```python
MINIMUM_LENGTH = 10

def get_h9_reviews(*data):
  h9_reviews = []
  for reviews in data:
    for _, row in reviews.iterrows():
      text = emoji.replace_emoji(str(row["text"]), replace='')
      unique_words = set(text.split())
      if len(unique_words) > MINIMUM_LENGTH and 'نقاط' not in unique_words:
        res = contains_cognitive(text, cognitive_phrases)
        if not res:
          h9_reviews.append(row["text"])

  return h9_reviews
```

```python
h9_reviews = get_h9_reviews(shein_data, afrd_data, company_data, balance_reviews)
```

```python
len(h9_reviews)
```

```python
h9_df = pd.DataFrame(h9_reviews, columns=["Text"])
h9_df.to_csv("/content/drive/MyDrive/New_data/Data/h9_reviews.csv", index=False)
```

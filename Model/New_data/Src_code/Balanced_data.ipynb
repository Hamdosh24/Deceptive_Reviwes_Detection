# Prepare

```python
import pandas as pd
```

```python
unbalanced_data = pd.read_csv("/content/drive/MyDrive/New_data/Data/unbalanced_new_data.csv")
```

# Balancing The Data

```python
# Hypothes balancing
min_ones = unbalanced_data[['H1', 'H2', 'H4', 'H5', 'H6', 'H7', 'H9', 'H10']].sum().min()
balanced_data = pd.DataFrame()
for column in ['H1', 'H2', 'H4', 'H5', 'H6', 'H7', 'H9', 'H10']:
    ones_data = unbalanced_data[unbalanced_data[column] == 1].sample(n=min_ones, random_state=42, replace=False)
    balanced_data = pd.concat([balanced_data, ones_data])

print(balanced_data[['H1', 'H2', 'H4', 'H5', 'H6', 'H7', 'H9', 'H10']].sum())
```

```python
len(balanced_data)
```

```python
balanced_data.to_csv("/content/drive/MyDrive/New_data/Data/balanced_hypothesis_data.csv", index=False)
```

```python
# polarity balancing
min_polarity = balanced_data['polarity'].value_counts().min()
balanced_polarity = pd.DataFrame()
for polarity in balanced_data['polarity'].unique():
    polarity_data = balanced_data[balanced_data['polarity'] == polarity].sample(n=min_polarity, random_state=42, replace=False)
    balanced_polarity = pd.concat([balanced_polarity, polarity_data])

print(balanced_polarity['polarity'].value_counts())
```

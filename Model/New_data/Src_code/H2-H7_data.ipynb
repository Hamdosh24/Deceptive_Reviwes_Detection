# H2

```python
MAX_LENGTH = 5
def get_h2_reviews(*data):
  h2_reviews = []
  for reviews in data:
    for _, row in reviews.iterrows():
      text = emoji.replace_emoji(str(row["text"]), replace='')
      unique_words = set(text.split())
      if len(unique_words) <= MAX_LENGTH:
        h2_reviews.append(row["text"])
  return h2_reviews
```

```python
h2_reviews = get_h2_reviews(shein_data, afrd_data, company_data, balance_reviews)
```

```python
len(h2_reviews)
```

```python
h2_df = pd.DataFrame(h2_reviews, columns=["Text"])
h2_df.to_csv("/content/drive/MyDrive/New_data/Data/h2_reviews.csv", index=False)
```

# H7

```python
MINIMUM_LENGTH = 30

def get_h7_reviews(*data):
  h7_reviews = []
  for reviews in data:
    for _, row in reviews.iterrows():
      text = emoji.replace_emoji(str(row["text"]), replace='')
      unique_words = set(text.split())
      if len(unique_words) > MINIMUM_LENGTH and 'نقاط' not in unique_words:
        h7_reviews.append(row["text"])
  return h7_reviews
```

```python
h7_reviews = get_h7_reviews(shein_data, afrd_data, company_data, balance_reviews)
```

```python
len(h7_reviews)
```

```python
h7_df = pd.DataFrame(h7_reviews, columns=["Text"])
h7_df.to_csv("/content/drive/MyDrive/New_data/Data/h7_reviews.csv", index=False)
```

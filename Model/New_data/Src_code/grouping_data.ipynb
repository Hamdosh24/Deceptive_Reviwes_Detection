# Grouping All the Data

```python
import pandas as pd
```

```python
# Deceptive
h1_data = pd.read_csv('/content/drive/MyDrive/New_data/Data/h1_reviews.csv')
h2_data = pd.read_csv('/content/drive/MyDrive/New_data/Data/h2_reviews.csv')
h4_data = pd.read_csv('/content/drive/MyDrive/New_data/Data/h4_reviews.csv')
h5_data = pd.read_csv('/content/drive/MyDrive/New_data/Data/h5_reviews.csv')

# Truthful
h6_data = pd.read_csv('/content/drive/MyDrive/New_data/Data/h6_reviews.csv')
h7_data = pd.read_csv('/content/drive/MyDrive/New_data/Data/h7_reviews.csv')
h9_data = pd.read_csv('/content/drive/MyDrive/New_data/Data/h9_reviews.csv')
h10_data = pd.read_csv('/content/drive/MyDrive/New_data/Data/h10_reviews.csv')
```

```python
h1_data['source'] = 'h1'
h2_data['source'] = 'h2'
h4_data['source'] = 'h4'
h5_data['source'] = 'h5'
h6_data['source'] = 'h6'
h7_data['source'] = 'h7'
h9_data['source'] = 'h9'
h10_data['source'] = 'h10'
```

```python
all_data = pd.concat([h1_data, h2_data, h4_data, h5_data, h6_data, h7_data, h9_data, h10_data], ignore_index=True)
```

```python
all_data = pd.get_dummies(all_data, columns=['source'])
```

```python
all_data_grouped = all_data.groupby('Text').agg({
    'source_h1': 'sum',
    'source_h2': 'sum',
    'source_h4': 'sum',
    'source_h5': 'sum',
    'source_h6': 'sum',
    'source_h7': 'sum',
    'source_h9': 'sum',
    'source_h10': 'sum'
}).reset_index()
```

```python
all_data_grouped = all_data_grouped.rename(columns={
    'source_h1': 'H1',
    'source_h2': 'H2',
    'source_h4': 'H4',
    'source_h5': 'H5',
    'source_h6': 'H6',
    'source_h7': 'H7',
    'source_h9': 'H9',
    'source_h10': 'H10'
})
```

```python
len(all_data_grouped)
```

```python
all_data_grouped['H1'] = all_data_grouped['H1'].apply(lambda x: 1 if x > 0 else 0)
all_data_grouped['H2'] = all_data_grouped['H2'].apply(lambda x: 1 if x > 0 else 0)
all_data_grouped['H4'] = all_data_grouped['H4'].apply(lambda x: 1 if x > 0 else 0)
all_data_grouped['H5'] = all_data_grouped['H5'].apply(lambda x: 1 if x > 0 else 0)
all_data_grouped['H6'] = all_data_grouped['H6'].apply(lambda x: 1 if x > 0 else 0)
all_data_grouped['H7'] = all_data_grouped['H7'].apply(lambda x: 1 if x > 0 else 0)
all_data_grouped['H9'] = all_data_grouped['H9'].apply(lambda x: 1 if x > 0 else 0)
all_data_grouped['H10'] = all_data_grouped['H10'].apply(lambda x: 1 if x > 0 else 0)
```

# Adding Polarity

```python
!pip install camel-tools
```

```python
from camel_tools.sentiment import SentimentAnalyzer

# Load the model and tokenizer
model_name = 'CAMeL-Lab/bert-base-arabic-camelbert-da-sentiment'
sa = SentimentAnalyzer(model_name,)


def get_polarity(text):
  predicted_class = sa.predict([text])
  return predicted_class[0]
```

```python
step = 50
for i in range(0,136501, step):
  print(f"stage: {int(1+(i/step))}/{int(135000/step)}")
  all_data_grouped.loc[i:i+step, 'polarity'] = all_data_grouped['Text'].iloc[i:i+step].apply(get_polarity)
  all_data_grouped.to_csv('/content/drive/MyDrive/New_data/Data/unbalanced_new_data.csv', index=False)
```

```python
exaggeration_phrases = [
    "اسوء",
    "اجمل",
    "افضل"
    "دائماً",
    "أبداً",
    "على الإطلاق",
    "لا يصدق",
    "بلا شك",
    "بلا منازع",
    "بلا حدود",
    "بلا مثيل",
    "لا مثيل له",
    "لا يضاهى",
    "لا يقدر بثمن",
    "لا يمكن تصديقه",
    "لا يمكن احتماله",
    "أعظم",
    "ولا غلطة",  # لهجة شامية
    "ولا بالأحلام",  # لهجة شامية
    "على آخر",  # لهجة شامية
    "حاجة تانية خالص",  # لهجة مصرية
    "فش حدا قدي",  # لهجة فلسطينية
    "ماكو مثله",  # لهجة عراقية
    "ولا بالأحلام",  # لهجة خليجية
    "شي خيالي",  # لهجة خليجية
    "ولا أروع",  # لهجة خليجية
    "ولع الدنيا",  # لهجة سودانية
    "بلا حدود",  # لهجة سودانية
    "بلا منازع",  # لهجة سودانية
    "مافي زيو",  # لهجة سودانية
    "أروع من الرائع",  # لهجة سودانية
    "كتير قوي",  # لهجة شامية
    "فشيخ",  # لهجة مصرية
    "خُرافي",  # لهجة خليجية
    "هبل",  # لهجة خليجية
    "خرافي",  # لهجة خليجية
    "يخبل",  # لهجة عراقية
    "سيء جدا",
    "اسوء شي",
    "جدا بشع",
    "في حياتي",
    "مافي متله",

]
```

```python
def contains_exaggeration(text, exaggeration_phrases):
  for phrase in exaggeration_phrases:
    if phrase in text:
      return True
  return False
```

# H1

```python
def get_h1_reviews(*data):
  h1_reviews = []
  for reviews in data:
    for _, row in reviews.iterrows():
      text = emoji.replace_emoji(str(row["text"]), replace='')
      res = contains_exaggeration(text, exaggeration_phrases)
      if res:
        h1_reviews.append(row["text"])
  return h1_reviews
```

```python
h1_reviews = get_h1_reviews(shein_data, afrd_data, company_data, balance_reviews)
```

```python
h1_df = pd.DataFrame(h1_reviews, columns=["Text"])
h1_df.to_csv("/content/drive/MyDrive/New_data/Data/h1_reviews.csv", index=False)
```

# H6

```python
MINIMUM_LENGTH = 15

def get_h6_reviews(*data):
  h6_reviews = []
  for reviews in data:
    for _, row in reviews.iterrows():
      text = emoji.replace_emoji(str(row["text"]), replace='')
      unique_words = set(text.split())
      if len(unique_words) > MINIMUM_LENGTH and 'نقاط' not in unique_words:
        res = contains_exaggeration(text, exaggeration_phrases)
        if not res:
          h6_reviews.append(row["text"])

  return h6_reviews
```

```python
h6_reviews = get_h6_reviews(shein_data, afrd_data, company_data, balance_reviews)
```

```python
h6_df = pd.DataFrame(h6_reviews, columns=["Text"])
h6_df.to_csv("/content/drive/MyDrive/New_data/Data/h6_reviews.csv", index=False)
```
